```python
from random import randint
import sys
MIN=30
MAX=1000
MULT=5
print(f"Guess a number between {MIN*MULT} and {MAX*MULT} that is a multiple of {MULT}")
a = randint(MIN,MAX)*MULT
b = int(input())
n=1
while a != b:
    if a > b:
        print("too low")
    elif a < b:
        print("too high")
    b = int(input())
    n = n + 1
print ("you win")
print (f"It took you {n} guesses")
```
