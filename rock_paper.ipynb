```python
print ("Welcome to rock paper scissors!")
user = int(input('Pick rock (1), paper (2), or scissors(3):'))
import random
robot = random.randint(1,3)

if robot == 1:
    print ('rock')
if robot == 2:
    print ('paper')
if robot == 3:
    print ('scissors')

if user == robot:
    print("It's a tie!")
if user == 1 and robot == 2:
    print("Computer wins")
if user == 2 and robot == 1:
    print("You won")
if user == 2 and robot == 3:
    print("Computer wins")
if user == 3 and robot == 2:
    print("You won")
if user == 3 and robot == 1:
    print("Computer wins")
if user == 1 and robot == 3:
    print("You won")
```
